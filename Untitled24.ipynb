LLM model finetune

```python
# Step 2: Format dataset for instruction fine-tuning
def format_instruction(example):
    instruction = example["instruction"]
    context = example["context"] if example["context"] else ""
    response = example["response"]
    return f"### Instruction:\n{instruction}\n\n### Context:\n{context}\n\n### Response:\n{response}"

# Apply formatting to dataset
llm_dataset = llm_dataset.map(lambda x: {"text": format_instruction(x)})
print("Formatted dataset sample:", llm_dataset[0]["text"])
```

```python
# Step 3: Load and quantize LLM
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16,
    bnb_4bit_use_double_quant=True
)

llm_model_name = "bigscience/bloomz-560m"
llm = AutoModelForCausalLM.from_pretrained(
    llm_model_name,
    quantization_config=bnb_config,
    device_map="auto",
    token=HF_TOKEN
)
llm_tokenizer = AutoTokenizer.from_pretrained(
    llm_model_name,
    token=HF_TOKEN
)
llm_tokenizer.pad_token = llm_tokenizer.eos_token  # Set pad token for batching
print(f"LLM ({llm_model_name}) loaded and quantized successfully.")
```

```python
from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training
# Step 4: Configure QLoRA for bloomz-560m
llm = prepare_model_for_kbit_training(llm)
lora_config = LoraConfig(
    r=16,  # Rank of LoRA adapters
    lora_alpha=16,  # Scaling factor
    target_modules=["query_key_value"],  # Correct module for BLOOM
    lora_dropout=0.05,  # Dropout for regularization
    bias="none",
    task_type="CAUSAL_LM"
)
llm = get_peft_model(llm, lora_config)
print("QLoRA configuration applied.")
```

```python
# Step 5: Set up training
training_args = TrainingArguments(
    output_dir="./bloomz-560m-finetuned",
    num_train_epochs=1,  # One pass over dataset
    per_device_train_batch_size=4,  # Small batch size for T4 GPU
    gradient_accumulation_steps=4,  # Accumulate gradients for effective larger batch size
    optim="paged_adamw_32bit",  # Optimizer for quantized models
    learning_rate=2e-4,
    max_grad_norm=0.3,  # Gradient clipping
    warmup_ratio=0.03,  # Warmup for learning rate
    lr_scheduler_type="cosine",
    logging_steps=10,
    save_strategy="epoch",
    fp16=True,  # Mixed precision training
)

trainer = SFTTrainer(
    model=llm,
    tokenizer=llm_tokenizer,
    train_dataset=llm_dataset,
    dataset_text_field="text",
    args=training_args,
    max_seq_length=512,  # Max sequence length for inputs
)
print("Trainer initialized.")
```

```python
# Step 6: Run fine-tuning
trainer.train()
print("Fine-tuning completed.")

# Step 7: Save fine-tuned LoRA adapters
llm.save_pretrained("./bloomz-560m-finetuned")
llm_tokenizer.save_pretrained("./bloomz-560m-finetuned")
print("Fine-tuned model and tokenizer saved to './bloomz-560m-finetuned'.")
```

Fine Tune embedding model

```python
from sentence_transformers import SentenceTransformer, InputExample, losses, evaluation
from datasets import load_dataset
from torch.utils.data import DataLoader
```

```python
# Step 1: Load dataset
embedding_dataset = load_dataset("sentence-transformers/stsb", split="train")
print("Embedding Dataset (STSbenchmark) loaded:", embedding_dataset)
print("Sample embedding dataset entry:", embedding_dataset[0])
```

```python
# Step 2: Format dataset for sentence-transformers
def format_stsb_example(example):
    return InputExample(texts=[example["sentence1"], example["sentence2"]], label=float(example["score"]) / 5.0)
```

```python
train_examples = [format_stsb_example(example) for example in embedding_dataset]
print("Formatted training examples sample:", train_examples[0].texts, train_examples[0].label)
```

```python
# Step 3: Load embedding model
embedding_model_name = "sentence-transformers/all-MiniLM-L6-v2"
embedding_model = SentenceTransformer(embedding_model_name, device=device)
print(f"Embedding model ({embedding_model_name}) loaded successfully.")
```

```python
# Step 4: Set up DataLoader and loss function
train_dataloader = DataLoader(train_examples, shuffle=True, batch_size=16)
train_loss = losses.CosineSimilarityLoss(model=embedding_model)
```

```python
# Step 5: Fine-tune the model
embedding_model.fit(
    train_objectives=[(train_dataloader, train_loss)],
    epochs=1,
    warmup_steps=100,
    optimizer_params={"lr": 2e-5},
    output_path="./all-MiniLM-L6-v2-finetuned",
    show_progress_bar=True
)
print("Fine-tuning completed.")
```

```python
# Step 6: Evaluate the model (optional)
# Load validation split for evaluation
eval_dataset = load_dataset("sentence-transformers/stsb", split="validation")
evaluator = evaluation.EmbeddingSimilarityEvaluator(
    sentences1=[ex["sentence1"] for ex in eval_dataset],
    sentences2=[ex["sentence2"] for ex in eval_dataset],
    scores=[float(ex["score"]) / 5.0 for ex in eval_dataset],
    batch_size=16,
    show_progress_bar=True
)
eval_score = embedding_model.evaluate(evaluator)
print(f"Validation Spearman correlation: {eval_score}")
```

```python
# Step 7: Save the fine-tuned model
embedding_model.save("./all-MiniLM-L6-v2-finetuned")
print("Fine-tuned embedding model saved to './all-MiniLM-L6-v2-finetuned'.")
```

```python
import torch
import faiss
import numpy as np
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
from peft import PeftModel
from sentence_transformers import SentenceTransformer
from langchain.document_loaders import WebBaseLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import FAISS
from langchain.chains import RetrievalQA
from langchain.prompts import PromptTemplate
from langchain.llms.huggingface_pipeline import HuggingFacePipeline
from langchain.embeddings import HuggingFaceEmbeddings
from langchain_community.docstore.in_memory import InMemoryDocstore

# Ensure CUDA is available
device = "cuda" if torch.cuda.is_available() else "cpu"

# Step 1: Load fine-tuned LLM with LoRA adapters (no merging)
llm_model_name = "bigscience/bloomz-560m"
base_model = AutoModelForCausalLM.from_pretrained(
    llm_model_name,
    torch_dtype=torch.bfloat16,
    device_map="auto"
)
llm = PeftModel.from_pretrained(
    base_model,
    "./bloomz-560m-finetuned",
    device_map="auto"
)
llm_tokenizer = AutoTokenizer.from_pretrained("./bloomz-560m-finetuned")
llm_tokenizer.padding_side = "left"
llm_tokenizer.pad_token = llm_tokenizer.eos_token

# Create a text-generation pipeline
llm_pipeline = pipeline(
    "text-generation",
    model=llm,
    tokenizer=llm_tokenizer,
    max_length=1024,
    truncation=True,
    device_map="auto",
    do_sample=True,
    top_p=0.95,      # Adjusted for more accurate outputs
    temperature=0.6  # Lowered for less randomness
)

# Step 2: Load fine-tuned embedding model
embedding_model = SentenceTransformer("./all-MiniLM-L6-v2-finetuned", device=device)
embedding_function = HuggingFaceEmbeddings(
    model_name="./all-MiniLM-L6-v2-finetuned",
    model_kwargs={"device": device}
)

# Step 3: Load web content
url = "http://jalammar.github.io/illustrated-transformer/"
try:
    loader = WebBaseLoader(url)
    documents = loader.load()
    if not documents:
        raise ValueError("No content loaded from the URL.")
except Exception:
    from langchain_core.documents import Document
    documents = [Document(page_content="Placeholder content for The Illustrated Transformer. Please replace with actual scraped content.", metadata={"source": url})]

# Step 4: Split documents into chunks
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=512,
    chunk_overlap=50,
    length_function=len
)
chunks = text_splitter.split_documents(documents)

# Step 5: Create vector store with FAISS
embeddings = embedding_model.encode([chunk.page_content for chunk in chunks], convert_to_numpy=True)
dimension = embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(embeddings)
docstore = InMemoryDocstore({str(i): chunk for i, chunk in enumerate(chunks)})
index_to_docstore_id = {i: str(i) for i in range(len(chunks))}
vector_store = FAISS(
    embedding_function=embedding_function,
    index=index,
    docstore=docstore,
    index_to_docstore_id=index_to_docstore_id
)

# Step 6: Set up RAG pipeline with LangChain
prompt_template = """### Instruction:
Use the following context to answer the query. Provide a concise and accurate response based only on the context, without reproducing the query or context in the output.

### Context:
{context}

### Query:
{question}

### Response:
"""
prompt = PromptTemplate(template=prompt_template, input_variables=["context", "question"])

# Wrap the LLM pipeline for LangChain
langchain_llm = HuggingFacePipeline(pipeline=llm_pipeline)

# Create RetrievalQA chain
qa_chain = RetrievalQA.from_chain_type(
    llm=langchain_llm,
    chain_type="stuff",
    retriever=vector_store.as_retriever(search_kwargs={"k": 3}),
    chain_type_kwargs={"prompt": prompt},
    return_source_documents=False  # Disable source documents in output
)

# Step 7: Test the RAG system
query = "What is the attention mechanism in transformers?"
result = qa_chain.invoke({"query": query})
print(result['result'])
```
